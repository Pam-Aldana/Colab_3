```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('/content/Variables.csv',sep=';')
```

```python
df.describe()
```

```python
df = df.drop(['ho_t_1','c_t_1', 'ht_t_1', 'ro', 'ho', 'c', 'ht'], axis=1)
```

```python
df.info()
```

```python
# Suponiendo que tienes un DataFrame llamado 'df'
df['categoria'] = df['categoria'].astype(str)
df['sexo'] = df['sexo'].astype(str)
df['grupo'] = df['grupo'].astype(str)
```

```python
# Convertir la columna 'ro_t_1' a números (float)
df['ro_t_1'] = pd.to_numeric(df['ro_t_1'], errors='coerce')

# Verificar si hay valores nulos después de la conversión
print(df['ro_t_1'].isnull().sum())  # Para contar valores nulos en la columna
```

```python
# Agrupar por Categoria, Sexo y Grupo y calcular la media de la remuneración
df = df.groupby(['categoria', 'sexo', 'grupo'])['ro_t_1'].mean().reset_index()
print(df)
```

```python
# Estadísticos descriptivos por Categoria, Sexo y Grupo para la remuneración
df2= df.groupby(['categoria', 'sexo', 'grupo'])['ro_t_1'].describe()
print(df2)
```
